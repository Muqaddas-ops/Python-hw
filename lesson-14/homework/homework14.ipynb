```python
import json

text = '''
{
    "students": [
      {
        "id": 1,
        "name": "Nodir",
        "age": 27,
        "pets": ["dog", "cat", "fish"],
        "familymemebers": [
          {"id": 1, "age": 25, "name": "alex", "relative": "Brother"},
          {"id": 2, "age": 21, "name": "sara", "relative": "Sister"},
          {"id": 3, "age": 22, "name": "Hamzahon", "relative": "Uncle"}
        ]
      }
    ]
}
'''

data = json.loads(text)

with open('family members.csv', 'w', encoding='utf-8') as f:
   
    f.write("student_name,member_id,member_name,age,relative\n")
    
    for student in data["students"]:
        for member in student["familymemebers"]:
            line = f"{student['name']},{member['id']},{member['name']},{member['age']},{member['relative']}\n"
            f.write(line)
```
